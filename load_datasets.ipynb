```python
#import dependencias
import pandas as pd
import pymysql.cursors
```

```python
# conectar ao banco de dados
def connect():
        con = pymysql.connect(
            host="",
            user="",
            database="",
            database="",
            password="",
            cursorclass=pymysql.cursors.DictCursor,
        )
        return con
```

```python
#Endereço do arquivo csv
csv_base = 'C:/ProgramData/MySQL/MySQL Server 8.0/Uploads/'  
csv_file_Original = f'${csv_base}dataset - original.csv'  
csv_file_Alterado = f'${csv_base}dataset.csv'
```

```python
def inserir_dados_original(con):
    """Insere dados do arquivo CSV original na tabela 'bookings'."""

    # Define a consulta SQL para o arquivo original
    insert_query_original = """
        INSERT INTO original (
            no_of_adults, no_of_children, no_of_weekend_nights, no_of_week_nights, 
            type_of_meal_plan, required_car_parking_space, room_type_reserved, lead_time, 
            arrival_year, arrival_month, arrival_date, market_segment_type, 
            repeated_guest, no_of_previous_cancellations, no_of_previous_bookings_not_canceled, 
            no_of_special_requests, booking_status, avg_price_per_room
        ) VALUES (%(no_of_adults)s, %(no_of_children)s, %(no_of_weekend_nights)s, 
                     %(no_of_week_nights)s, %(type_of_meal_plan)s, 
                     %(required_car_parking_space)s, %(room_type_reserved)s, %(lead_time)s, 
                     %(arrival_year)s, %(arrival_month)s, %(arrival_date)s, 
                     %(market_segment_type)s, %(repeated_guest)s, 
                     %(no_of_previous_cancellations)s, %(no_of_previous_bookings_not_canceled)s, 
                     %(no_of_special_requests)s, %(booking_status)s, %(avg_price_per_room)s)
    """

    try:
        with con.cursor() as cursor:
            # Carrega o arquivo CSV original
            data_original = pd.read_csv(csv_file_Original)

            # Verifica se as colunas do DataFrame estão corretas
            if set(data_original.columns) != set(insert_query_original.split('(')[1].split(')')[0].replace(' ', '').split(',')):
                raise ValueError("As colunas do DataFrame não correspondem às colunas da tabela")

            # Insere os dados em massa
            data_to_insert = [
                {
                    'no_of_adults': row['no_of_adults'],
                    'no_of_children': row['no_of_children'],
                    'no_of_weekend_nights': row['no_of_weekend_nights'],
                    'no_of_week_nights': row['no_of_week_nights'],
                    'type_of_meal_plan': row['type_of_meal_plan'],
                    'required_car_parking_space': row['required_car_parking_space'],
                    'room_type_reserved': row['room_type_reserved'],
                    'lead_time': row['lead_time'],
                    'arrival_year': row['arrival_year'],
                    'arrival_month': row['arrival_month'],
                    'arrival_date': row['arrival_date'],
                    'market_segment_type': row['market_segment_type'],
                    'repeated_guest': row['repeated_guest'],
                    'no_of_previous_cancellations': row['no_of_previous_cancellations'],
                    'no_of_previous_bookings_not_canceled': row['no_of_previous_bookings_not_canceled'],
                    'no_of_special_requests': row['no_of_special_requests'],
                    'booking_status': row['booking_status'],
                    'avg_price_per_room': row['avg_price_per_room']
                }
                for index, row in data_original.iterrows()
            ]
            cursor.executemany(insert_query_original, data_to_insert)
            con.commit()
            print("Dados do arquivo original inseridos com sucesso!")

    except :
        print("Erro na inserção de dados do arquivo original")
        con.rollback()
```

```python
def inserir_dados_alterado(con):
    """Insere dados do arquivo CSV alterado na tabela 'bookings'."""

    # Define a consulta SQL para o arquivo alterado
    insert_query_alterado = """
        INSERT INTO reservations (
            label_avg_price_per_room, no_of_adults, no_of_children, 
            no_of_weekend_nights, no_of_week_nights, type_of_meal_plan, 
            required_car_parking_space, room_type_reserved, lead_time
        ) VALUES (%(label_avg_price_per_room)s, %(no_of_adults)s, %(no_of_children)s, 
                     %(no_of_weekend_nights)s, %(no_of_week_nights)s, %(type_of_meal_plan)s, 
                     %(required_car_parking_space)s, %(room_type_reserved)s, %(lead_time)s)
    """

    try:
        with con.cursor() as cursor:
            # Carrega o arquivo CSV alterado
            data_alterado = pd.read_csv(csv_file_Alterado)

            # Verifica se as colunas do DataFrame estão corretas
            if set(data_alterado.columns) != set(insert_query_alterado.split('(')[1].split(')')[0].replace(' ', '').split(',')):
                raise ValueError("As colunas do DataFrame não correspondem às colunas da tabela")

            # Insere os dados em massa
            data_to_insert = [
                {
                    'label_avg_price_per_room': row['label_avg_price_per_room'],
                    'no_of_adults': row['no_of_adults'],
                    'no_of_children': row['no_of_children'],
                    'no_of_weekend_nights': row['no_of_weekend_nights'],
                    'no_of_week_nights': row['no_of_week_nights'],
                    'type_of_meal_plan': row['type_of_meal_plan'],
                    'required_car_parking_space': row['required_car_parking_space'],
                    'room_type_reserved': row['room_type_reserved'],
                    'lead_time': row['lead_time']
                }
                for index, row in data_alterado.iterrows()
            ]
            cursor.executemany(insert_query_alterado, data_to_insert)
            con.commit()
            print("Dados do arquivo alterado inseridos com sucesso!")

    except :
        print("Erro na inserção de dados do arquivo alterado:")
        con.rollback()
```

```python
con = connect()
inserir_dados_original(con)
inserir_dados_alterado(con)
```
